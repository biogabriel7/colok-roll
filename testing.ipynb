```python
from perinuclear_analysis.imaging_preprocessing.background_subtraction import subtract_background_percentile_roi

# Inputs:
img_zyxc: np.ndarray of shape (Z, Y, X, C)
mask_2d:  np.ndarray of shape (Y, X), labeled or boolean
# For labeled masks, turn into union-of-labels:
roi_union_2d = (mask_2d > 0)

# Apply ROI-based percentile background subtraction (per-channel)
corrected = subtract_background_percentile_roi(img_zyxc, roi_union_2d, percentile=1.0)

# Quick visualize one z-slice and one channel
z = img_zyxc.shape[0] // 2
ch = 0
vmin, vmax = np.percentile(img_zyxc[z, :, :, ch], [1, 99])

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(img_zyxc[z, :, :, ch], cmap="magma", vmin=vmin, vmax=vmax)
axes[0].set_title("Original")
axes[0].axis("off")

vmin2, vmax2 = np.percentile(corrected[z, :, :, ch], [1, 99])
axes[1].imshow(corrected[z, :, :, ch], cmap="magma", vmin=vmin2, vmax=vmax2)
axes[1].set_title("Corrected (p=1.0%)")
axes[1].axis("off")
plt.show()
```
